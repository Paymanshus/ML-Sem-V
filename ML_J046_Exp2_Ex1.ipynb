```
filename = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
df = pd.read_csv(filename)
```

###Adding Column Headers

```
df.columns = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
"drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
"num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
"peak-rpm","city-mpg","highway-mpg","price"]
```

###Checking for na, missing and incorrect values

```
df.isna().sum()
```

```
df.dtypes
```

```
qindices = df[pd.isna(pd.to_numeric(df['normalized-losses'],errors='coerce'))].index.tolist()
qindices
```

This gives us a list of invalid entries ie '?' values in a column whose data type should be numbers

###Replace '?' with NaN

```
df.replace('?',np.nan,inplace=True)
df
```

###Count missing values in each column and print it with column name

```
df.isnull().values
```

```
null_count = {}
for i in df.columns:
  null_count[i]=df[i].isnull().values.sum()
```

```
null_count
```

##Deleting rows with missing data

```
df.shape
```

```
df.dropna(axis=0,subset=['price'], inplace = True)
df.shape
```

```
df
```

Replacing missing with mean of column

```
# df.fillna(df.mean(), inplace = True)
# df
for i in ['normalized-losses','bore','stroke','horsepower','peak-rpm']:
  df[i] = df[i].fillna((df[i].mean()))
```
